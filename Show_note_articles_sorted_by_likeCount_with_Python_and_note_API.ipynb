```
import requests
import pandas as pd
import json
import time
from tqdm import tqdm

# Fill in note user name
# https://note.com/XXX
user_name="XXX"

# note API
user_api_url="https://note.com/api/v2/creators/%s"%user_name
contents_api_url="https://note.com/api/v2/creators/%s/contents"%user_name

# Get user info
res = requests.get(user_api_url).content
df = pd.read_json(res)

# 取得秒数間隔の指定
interval = 1

# Get note_count
note_count = df['data']['noteCount']

# 総取得ページ数
page_num = (note_count//6) + 1

# 個別ページの取得
df2 = pd.DataFrame()

keys = ["name", "likeCount", "noteUrl", "publishAt"]

for page in tqdm(range(page_num)):

   page = page + 1

   payload = {'kind':'note', 'page':page}
   res = requests.get(contents_api_url, params=payload).content

   df = pd.read_json(res)           

   df_notes = []
   for i in range(len(df["data"]["contents"])):
     d = {k: v for k, v in df["data"]["contents"][i].items() if k in keys}
     df_notes.append(d)
   
   # 展開して新しいデータフレームに行として追加していく
   for i in range(len(df_notes)):
       df_n = pd.DataFrame(df_notes[i].values(), index=df_notes[i].keys()).T
       df2 = pd.concat([df2, df_n])

   time.sleep(interval)
   
# 重複行を削除する
df2.drop_duplicates(subset='noteUrl', inplace=True)

# like数順に並び替え
df2 = df2.sort_values('likeCount', ascending=False)

# インデックスの振り直し
df2.reset_index(drop=True, inplace=True)
df2.index = df2.index + 1

# Convert to a DataFrame and render.
pd.set_option('display.max_rows', 100)
pd.DataFrame.from_dict(df2)


```
